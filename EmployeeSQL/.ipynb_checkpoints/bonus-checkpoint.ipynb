```python
# Dependencies
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
engine = create_engine('postgresql+psycopg2://postgres:postgres@localhost:5432/SQL_HW')
connection = engine.connect()
```

```python
employees = pd.read_sql('SELECT * FROM employees', connection)
employees
```

```python
salaries = pd.read_sql('SELECT * FROM salaries', connection)
salaries
```

```python
titles = pd.read_sql('SELECT * FROM titles', connection)
titles
```

```python
titles1 = titles.rename(columns={'title_id':'emp_title_id'})
titles1
```

```python
emp_sal = pd.merge(employees, salaries, on="emp_no")
emp_sal
```

```python
merged_df = pd.merge(emp_sal, titles1, on='emp_title_id')
merged_df
```

```python
plt.hist(merged_df['salary'])
plt.xlabel("Salary")
plt.ylabel('Number of Employees')
plt.title('Frequency of Salaries')
plt.show()
```

```python
staff = merged_df.loc[merged_df['title'] == 'Staff']
staff_sal = staff['salary'].mean()
staff_sal
```

```python
ass_eng = merged_df.loc[merged_df['title'] == 'Assistant Engineer']
ass_eng_sal = ass_eng['salary'].mean()
ass_eng_sal
```

```python
sen_staff = merged_df.loc[merged_df['title'] == 'Senior Staff']
sen_staff_sal = sen_staff['salary'].mean()
sen_staff_sal
```

```python
eng = merged_df.loc[merged_df['title'] == 'Engineer']
eng_sal = eng['salary'].mean()
eng_sal
```

```python
sen_eng = merged_df.loc[merged_df['title'] == 'Senior Engineer']
sen_eng_sal = sen_eng['salary'].mean()
sen_eng_sal
```

```python
tech = merged_df.loc[merged_df['title'] == 'Technique Leader']
tech_sal = tech['salary'].mean()
tech_sal
```

```python
mgr = merged_df.loc[merged_df['title'] == 'Manager']
mgr_sal = mgr['salary'].mean()
mgr_sal
```

```python
title_name = ['Staff', 'Senior Staff', 'Assistant Engineer', 'Engineer', 'Senior Engineer', 'Technique Leader', 'Manager']
avg_sal = [staff_sal, sen_staff_sal, ass_eng_sal, eng_sal, sen_eng_sal, tech_sal, mgr_sal]
```

```python
plt.bar(title_name, avg_sal)
plt.xticks(title_name, rotation='vertical')
plt.xlabel('Job Title')
plt.ylabel('Average Salary')
plt.title('Average Salary by Job Title')
plt.show()
```
